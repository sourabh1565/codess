```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
```

```python
air=pd.read_csv('https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/AirPassengers/AirPassengers.csv')
air.head(10)
```

```python
air.columns.values
```

```python
air['Passengers'].corr(air['Promotion_Budget'])
```

```python
(air.Passengers).corr(air.Promotion_Budget)
```

```python
np.corrcoef(air.Passengers,air.Promotion_Budget)
```

```python
air_model=sm.ols(formula='Passengers ~ air.Promotion_Budget',data=air)
fit1=air_model.fit()
fit1.summary()
```

```python
air_model2=sm.ols(formula='Passengers ~ Inter_metro_flight_ratio',data=air)
fit2=air_model2.fit()
fit2.summary()
```

```python
air_model3=sm.ols(formula='Passengers ~  air.Promotion_Budget + Inter_metro_flight_ratio + Service_Quality_Score',data=air)
fit3=air_model3.fit()
fit3.summary()
```

```python
#Code for VIF Calculation
#Writing a function to calculate the VIF values

def vif_cal(input_data):
    x_vars = input_data
    xvar_names=x_vars.columns
    for i in range(0,xvar_names.shape[0]):
        y=x_vars[xvar_names[i]]
        x=x_vars[xvar_names.drop(xvar_names[i])]
        rsq=sm.ols(formula="y~x", data=x_vars).fit().rsquared
        vif=round(1/(1-rsq),2)
        print (xvar_names[i], " VIF = " , vif)
```
